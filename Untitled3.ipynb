```python

from google.colab import files
import pandas as pd
uploaded = files.upload()
df = pd.read_csv("E-commerce Dataset.csv")
df.head()
```

```python
print(df.isnull().sum())
df = df.dropna()
df = df.drop_duplicates()
print("Remaining rows:", len(df))
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df['order_date'] = pd.to_datetime(df['order_date'])
df['time'] = pd.to_datetime(df['time'], errors='coerce')
df['customer_id'] = df['customer_id'].astype(str)
df['gender'] = df['gender'].astype('category')
df['device_type'] = df['device_type'].astype('category')
df['customer_login_type'] = df['customer_login_type'].astype('category')
df['product_category'] = df['product_category'].astype('category')
df['product'] = df['product'].astype(str)
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce')
df['discount'] = pd.to_numeric(df['discount'], errors='coerce')
df['profit'] = pd.to_numeric(df['profit'], errors='coerce')
df['shipping_cost'] = pd.to_numeric(df['shipping_cost'], errors='coerce')
df['order_priority'] = df['order_priority'].astype('category')
df['payment_method'] = df['payment_method'].astype('category')
df.info()
```

```python
df['margin'] = df['profit'] / df['sales']
df['high_value_customer'] = df['sales'].apply(lambda x: 1 if x > 1000 else 0)
df['discount_flag'] = df['discount'].apply(lambda x: 1 if x > 0 else 0)
df['priority_flag'] = df['order_priority'].apply(lambda x: 1 if x == 'High' else 0)
df[['margin','high_value_customer','discount_flag','priority_flag']].head()
```

```python
customers_df = df[['customer_id','gender','device_type','customer_login_type']].drop_duplicates()
orders_df = df[['order_date','time','customer_id','sales','quantity',
                'discount','profit','shipping_cost','order_priority',
                'payment_method','margin','high_value_customer',
                'discount_flag','priority_flag']]
products_df = df[['product','product_category','sales','quantity','profit','margin']]

print("Customers:", customers_df.shape)
print("Orders:", orders_df.shape)
print("Products:", products_df.shape)
```

```python

customers_df.to_csv("customers.csv", index=False)
orders_df.to_csv("orders.csv", index=False)
products_df.to_csv("products.csv", index=False)

print("Export complete: customers.csv, orders.csv, products.csv")
```
